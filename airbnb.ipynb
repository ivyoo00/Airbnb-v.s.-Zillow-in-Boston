```python
# import
import requests
import time
import json
from datetime import datetime
import pandas as pd
import numpy as np
import csv
import math
```

```python
file = "/Users/gejiali/study/ECE/Spring2018/EC500/project/listings.csv"
df = pd.read_csv(file)
df = df[["id", "neighbourhood", "bathrooms", "beds", "price", "review_scores_rating",
         "review_scores_cleanliness", "review_scores_location", "latitude", "longitude"]]

df.fillna(0, inplace=True)
li = np.array(df)
data = li.tolist()
price = []
for k in range(len(data)):
    pr = data[k][4].split(".")[0][1:]
    if len(pr) >= 5:
        p = int(pr.split(",")[0] + pr.split(",")[1])
    else:
        p = int(pr)
        
    if (data[k][3] == 0 or math.isnan(float(data[k][3])) == True):
        price.append(float(p / 1))
    else:
        price.append(float(p / data[k][3]))

num = len(price)
s_price = sorted(price)
print("price is classified by:", s_price[int(num * 0.2)], s_price[int(num * 0.4)],
       s_price[int(num * 0.6)], s_price[int(num * 0.8)])
                     
```

```python
infor = []
for i in range(len(data)):
    dic = dict()
    dic.setdefault("id", data[i][0])
    dic["neighborhood"] = data[i][1]
    if data[i][3] == 0 or math.isnan(float(data[i][3])) == True:
        dic["bathrooms/bed"] = float(data[i][2] / 1)
    else:
        dic["bathrooms/bed"] = float(data[i][2] / data[i][3])
    dic["reviewscore_rating"] = data[i][5]
    dic["score_cleanliness"] = data[i][6]
    dic["score_location"] = data[i][7]
    dic["latitude"] = data[i][8]
    dic["longitude"] = data[i][9]
    dic["price"] = price[i]
    if price[i] <= 54:
        dic["price_class"] = 1
    elif price[i] <= 75:
        dic["price_class"] = 2
    elif price[i] <= 100:
        dic["price_class"] = 3
    elif price[i] <= 150:
        dic["price_class"] = 4
    else:
        dic["price_class"] = 5    
    infor.append(dic)
print(infor)
```

```python
field_names = ["id", "neighborhood", "bathrooms/bed", "reviewscore_rating",
               "score_cleanliness", "score_location", "latitude", "longitude", "price", "price_class"]
with open("/Users/gejiali/study/ECE/Spring2018/EC500/project/airbnb1.csv", "w") as f:
    writer = csv.DictWriter(f, field_names)
    writer.writerow({x: x for x in field_names})
    for item in infor:
        writer.writerow(item)
```

```python
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer()

vec.fit_transform(infor).toarray()
print(vec.get_feature_names())
print(len(vec.get_feature_names()))
print(vec.fit_transform(infor).toarray())
```
